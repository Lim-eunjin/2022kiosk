```
class InferenceConfig(coco.CocoConfig):
    # Set batch size to 1 since we'll be running inference on
    # one image at a time. Batch size = GPU_COUNT * IMAGES_PER_GPU
    GPU_COUNT = 1
    IMAGES_PER_GPU = 1

config = InferenceConfig()
```

```
#한번에 여러개 object detection 하기
#크롤링 데이터를 한번에 많이 돌리면 colab이 중간에 멈추므로 a,b변수를 통해 범위를 지정(1000개씩이 가장 best)

"""
  Runs the detection pipeline.

  images: List of images, potentially of different sizes.

  Returns a list of dicts, one dict per image. The dict contains:
  rois: [N, (y1, x1, y2, x2)] detection bounding boxes
  class_ids: [N] int class IDs
  scores: [N] float probability scores for the class IDs
  masks: [H, W, N] instance binary masks
"""
#범위를 여러번에 나눠 돌릴 것이므로, 변수 초기화
r1 = []
r2 = []
r3 = []

#데이터 돌리는 범위
a = 4001
b = 5001

for i in range(len(df['thumbnail'][a:b])) :
  #path2는 썸네일 이미지 경로 path2 = 'https://i.ytimg.com/vi/6JKKVtC-sTk/mqdefault.jpg'
  path2 = df['thumbnail']
  print(i+a, end=' ')

  try:
    image = skimage.io.imread(path2[i+a])
  
    # Run detection
    results = model.detect([image], verbose=0)

    # Visualize results
    r = results[0]
    r1.append(r['rois'])
    r2.append(r['class_ids'])
    r3.append(r['scores'])      

  #썸네일 url 주소 이상 등으로 이미지를 가져오지 못하는 경우 0으로 처리하고 계속 진행    
  except Exception:
    pass
    r1.append(np.array([[0,0,0,0]]))
    r2.append(np.array([99]))
    r3.append(np.array([0]))
    
  
```
